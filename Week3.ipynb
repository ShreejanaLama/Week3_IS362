Biking from NYC to Key West

This notebook tracks my cumulative and daily miles from New York City to Key West, Florida, using Pandas. It also includes a visualization of Key West.

```python
import pandas as pd

# Create a Pandas Series with cumulative miles recorded each day.
cumulative_miles = pd.Series([55, 120, 150, 190, 220, 270, 300, 330])  # Example data

# Calculate daily miles by finding the difference
daily_miles = cumulative_miles.diff().fillna(cumulative_miles.iloc[0])

# Create a DataFrame for better presentation
df = pd.DataFrame({'Cumulative Miles': cumulative_miles, 'Daily Miles': daily_miles})

# Display the DataFrame
print(df)
```
